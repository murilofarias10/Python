```python
# ============================================
# Decision Tree Classifier on iris.csv
# (Fixed columns: 4 features + target 'Species')
# ============================================

# Step 1: Import required libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_graphviz, export_text
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import graphviz
#from google.colab import files

# Step 2: Upload iris.csv file
#print("Please upload your iris.csv file")
#uploaded = files.upload()

# Step 3: Read uploaded file
#file_name = list(uploaded.keys())[0]
data = pd.read_csv('iris.csv')
#data = pd.read_csv(file_name)
print("\nFirst 5 rows:\n", data.head())

# Step 4: Define predictors and target
predictors = ['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width']
target_col = 'Species'


# Step 5: Separate features (X) and target (y)
X = data[predictors]
y = data[target_col]

# Step 6: Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Step 7: Train Decision Tree with depth=10 and leaf size=5
model = DecisionTreeClassifier(
    random_state=42,
    max_depth=10,
    min_samples_leaf=5
)
model.fit(X_train, y_train)

# Step 8: Make predictions
y_pred = model.predict(X_test)

# Step 9: Evaluate performance
print("\n Model Accuracy:", accuracy_score(y_test, y_pred))

# Step 10: Visualize Decision Tree
plt.figure(figsize=(16,10))
plot_tree(
    model,
    filled=True,
    feature_names=predictors,
    class_names=[str(c) for c in model.classes_],
    rounded=True,
    fontsize=10
)
plt.title("Decision Tree of Iris Dataset", fontsize=16)
plt.show()

# Step 11: Save tree as PNG
dot_data = export_graphviz(
    model,
    out_file=None,
    feature_names=predictors,
    class_names=[str(c) for c in model.classes_],
    filled=True,
    rounded=True,
    special_characters=True
)
graph = graphviz.Source(dot_data)
graph.render("iris_decision_tree_depth_leaf", format="png", cleanup=True)

print("\n The decision tree has been saved as 'iris_decision_tree_depth_leaf.png'.")

# Step 12: Download PNG
files.download("iris_decision_tree_depth_leaf.png")

# ============================================
# Step 13: Print Decision Rules from the Tree
# ============================================
rules = export_text(model, feature_names=predictors)
print("\n Decision Tree Rules:\n")
print(rules)
```

```python
X = data[predictors]
y = data[target_col]
```

```python
# Step 7: Train Decision Tree
model = DecisionTreeClassifier(
    random_state=42,
    max_depth=10,
    min_samples_leaf=5
)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# Step 9: Evaluate performance
print("\nModel Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Step 10: Visualize Decision Tree
plt.figure(figsize=(16,10))
plot_tree(
    model,
    filled=True,
    feature_names=predictors,
    class_names=[str(c) for c in model.classes_],
    rounded=True,
    fontsize=10
)
plt.show()
```
