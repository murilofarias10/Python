```python
#Question 2: Decision Tree Modelling
"""
2.1
Prepare training and testing sets with a 75-25 split ratio. 
This helps in machine learning, with 75% of the data used for training and 25% for testing.
"""
```

```python
#libraries to use on decision tree model

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
#first splitting my data into features and target
X = mu_df.drop("target", axis=1)   # all features
y = mu_df["target"]                # target column
```

```python
# rain/test split (75% train, 25% test)
#based on my student ID CT1004837 i will use random_state = 37
#stratify = yto keep the proportion of classes in both train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=37, stratify=y 
)
```

```python
#Question 2: Decision Tree Modelling
"""
2.2
Perform classification using the Decision Tree model on the split dataset.

Use random_state = the last two digits of your student ID. For example: 
If your student ID ends in 45, set random_state = 45.
Using the wrong random_state will result in a point deduction.
"""
```

```python
# 4) Decision Tree model
clf = DecisionTreeClassifier(
    criterion="gini",
    max_depth=4,          # keeps the tree readable
    min_samples_leaf=5,
    random_state=37
)
clf.fit(X_train, y_train)
```

```python
#Question 2: Decision Tree Modelling
"""
2.3
Calculate the accuracy of the Decision Tree Classifier to measure how well 
it performs on both the Training and Testing datasets
"""
```

```python
# 5) Evaluate
y_pred = clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
#model predicted correctly 84.4% of the wines’ classes.
```

```python
print("\nClassification report:\n", classification_report(y_test, y_pred))
```

```python
#Question 2: Decision Tree Modelling
"""
2.4
Show the confusion matrix to present the results of the Decision Tree Classifier.
"""
```

```python
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
plt.title("Confusion Matrix")
plt.show()
```

```python
"""
Confusion matrix:
 [[14  1  0] based on class 0
 [ 2 14  2] based on class 1
 [ 0  2 10]] based on class 2

class 0:
    14 wines of class 0 were correctly predicted as 0
    1 wine of class 0 was misclassified as class 1
    0 wines of class 0 were misclassified as class 2

class 1:
    14 wines of class 1 were correctly predicted as 1
    2 wines of class 1 were misclassified as class 0
    2 wines of class 1 were misclassified as class 2

class 2:
    0 wines of class 2 were misclassified as class 0
    2 wines of class 2 were misclassified as class 1
    10 wines of class 2 were correctly predicted as 2

 """
```

```python
#Visualize tree
import numpy as np

plt.figure(figsize=(18, 10))
plot_tree(
    clf,
    feature_names=X.columns,
    class_names=[str(c) for c in np.unique(y_train)],
    filled=True,
    rounded=True,
    fontsize=8
)
plt.title("Decision Tree predicting Wine Class (target)")
plt.show()
```
