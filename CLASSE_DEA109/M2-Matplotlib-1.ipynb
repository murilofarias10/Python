# Matplotlib
- Matplotlib is a widely used Python library for creating visualizations in data science, machine learning, and scientific computing
- It provides a versatile and comprehensive set of tools for creating a wide range of charts, graphs, plots, and other data visualizations.
- Whether we are exploring data, presenting results, or creating publication-quality plots, Matplotlib is an essential library for data scientist or analyst.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Line Plot

```python
# Cell 3: 
# Let's start by creating a simple line plot.

# Data
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 14, 8, 15, 20])

# Create a line plot
# We can use figsize to guarantee the figure has a certain size (in inches) and aspect ratio if saved to disk
plt.figure(figsize=(8, 4))
plt.plot(x, y, marker='o', linestyle='-')

plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

# Display the plot
plt.show()
```

## Scatter Plot

```python
# Now, let's create a scatter plot using random data points.

# Generate random data
np.random.seed(42)
x = np.random.rand(10)
y = np.random.rand(10)
colors = np.random.rand(10)
sizes = 1000 * np.random.rand(10)

print("x: ", x)
print("y: ", y)

# Create a scatter plot
# c=colors: This parameter specifies the color of each data point on the scatter plot
# s=sizes: This parameter determines the size of each data point on the scatter plot
# alpha=0.5: This parameter controls the transparency of the points in the scatter plot
# cmap='viridis': This parameter specifies the color map to be used for coloring the points based on the colors array
plt.figure(figsize=(8, 4))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='winter') 

plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

# Display the plot
plt.show()
```

## Bar Chart

```python
# Let's create a bar chart to visualize data.

# Data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [40, 60, 75, 90]

# Create a bar chart
plt.figure(figsize=(8, 4))
plt.bar(categories, values, color='skyblue')

plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')

# Display the plot
plt.show()
```

## Histogram

```python
# Generate random data
# randn(1000): This code calls the randn function from the np.random module
# The randn function generates random numbers from a standard normal distribution
data = np.random.randn(1000)

# Create a histogram
# bins=20: The bins parameter determines the number of intervals (or "bins") into which the data will be divided
# edgecolor='black': The edgecolor parameter determines the color of the edges or borders of the bars in the histogram
plt.figure(figsize=(8, 4))
plt.hist(data, bins=20, color='lightgreen', edgecolor='black')

plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

## Pie Chart

```python
# Data
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [30, 20, 25, 45]

# Create a pie chart
# autopct='%.1f%%': The autopct parameter controls the format of the percentage labels displayed on each segment of the pie chart
# startangle=90: The startangle parameter sets the angle at which the pie chart starts. 
plt.pie(sizes, labels=labels, autopct='%.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'lightcoral', 'lightsalmon'])
plt.title('Pie Chart')

# Display the plot
plt.show()
```

## Box Plots

```python
# Generate random data for three categories
# The normal function generates random numbers from a normal distribution.
  # The first argument, 0, represents the mean (average) of the normal distribution
  # The second argument, 1, represents the standard deviation of the normal distribution
  # The third argument, 100, specifies the number of random numbers to generate
data = [np.random.normal(0, 1, 100),  # Category 1
        np.random.normal(0, 2, 100),  # Category 2
        np.random.normal(1, 1, 100)]  # Category 3

# Create a box plot to visualize the data distribution.

# Create a figure and axis
plt.figure(figsize=(8, 4))
ax = plt.subplot(111)

# Create the box plot
# vert=True: This parameter specifies the orientation of the box plot. When vert=True, the box plot is oriented vertically
# patch_artist=True: The patch_artist parameter, when set to True, makes the individual components of the box plot 
  # (boxes, whiskers, caps, etc.) appear as artists that can be customized
boxplot = ax.boxplot(data, vert=True, patch_artist=True)

# Customize the box plot colors
colors = ['lightblue', 'lightgreen', 'lightcoral']
# Initiate a loop that iterates through pairs of the boxplot['boxes'] and the colors list
for patch, color in zip(boxplot['boxes'], colors): 
    patch.set_facecolor(color)

# Set labels for the x-axis
ax.set_xticklabels(['Category 1', 'Category 2', 'Category 3'])

# Set the title and labels
ax.set_title('Box Plot')
ax.set_xlabel('Categories')
ax.set_ylabel('Values')

# Display the box plot
plt.show()
```

## Subplots
- we will explore how to create and work with subplots in Matplotlib
- Subplots allow us to display multiple plots within a single figure, which is useful for comparing different visualizations or data

```python
fig = plt.figure(figsize=(8, 6))

# Subplot 1
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot([1, 2, 3, 4], [1, 4, 9, 16])
ax1.set_title('Subplot 1')

# Subplot 2
ax2 = fig.add_subplot(2, 2, 2)
ax2.scatter([1, 2, 3, 4], [1, 2, 3, 4], color='red')
ax2.set_title('Subplot 2')

# Subplot 3
ax3 = fig.add_subplot(2, 2, 3)
ax3.bar(['A', 'B', 'C', 'D'], [10, 20, 15, 30], color='green')
ax3.set_title('Subplot 3')

# Subplot 4
ax4 = fig.add_subplot(2, 2, 4)
ax4.hist([3, 4, 5, 6, 6, 6, 7, 7, 8, 9], bins=5, color='purple')
ax4.set_title('Subplot 4')

# Adjust layout
plt.tight_layout()

# Display the figure
plt.show()
```

## Sharing of Axis

```python
# Create a figure with two subplots sharing the x-axis
fig, axs = plt.subplots(2, 1, sharex=True, figsize=(6, 4))

# Subplot 1
axs[0].plot([1, 2, 3, 4], [1, 4, 9, 16], marker='o', color='blue')
axs[0].set_title('Subplot 1')
axs[0].set_ylabel('Y-axis for Subplot 1')

# Subplot 2
axs[1].scatter([1, 2, 3, 4], [0, 3, 2, 4], marker='x', color='red')
axs[1].set_title('Subplot 2')
axs[1].set_xlabel('X-axis')
axs[1].set_ylabel('Y-axis for Subplot 2')

# Adjust layout
plt.tight_layout()

# Display the figure
plt.show()
```

## Creating a Plot with Different Chart Types

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 4))

# Line plot
ax.plot(x, y, label='Line Plot', marker='o', linestyle='-')

# Bar plot
ax.bar(x, y, width=0.3, label='Bar Plot', align='center', alpha=0.6, color='green')

# Scatter plot
ax.scatter(x, y, label='Scatter Plot', color='red', marker='x', s=50)

# Set title and labels
ax.set_title('Different Chart Types in the Same Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Add a legend
ax.legend()

# Display the plot
plt.show()
```

```python
# Sample data for the first line chart
x1 = [1, 2, 3, 4, 5]
y1 = [10, 15, 13, 17, 20]

# Sample data for the second line chart
x2 = [1, 2, 3, 4, 5]
y2 = [5, 8, 16, 2, 15]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 4))

# First line chart
ax.plot(x1, y1, label='Line Chart 1', marker='o', linestyle='-', color='blue')

# Second line chart
ax.plot(x2, y2, label='Line Chart 2', marker='s', linestyle='--', color='red')

# Set title and labels
ax.set_title('Two Line Charts in the Same Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Add a legend
ax.legend(loc='best')

# Display the plot
plt.show()
```

## Plotting from Pandas

```python
import pandas as pd

iris_data = pd.read_csv("iris.csv")

# Display the first few rows of the DataFrame
print(iris_data.head())

# Create a scatter plot of 'sepal_length' vs. 'sepal_width'
plt.figure(figsize=(8, 4))
plt.scatter(iris_data['sepal_length'], iris_data['sepal_width'])

# Customize the plot
plt.title("Scatter Plot of Sepal Length vs. Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")

# Display the plot
plt.show()
```
