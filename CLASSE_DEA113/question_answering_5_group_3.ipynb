```python

qa = pipeline("question-answering")
```

This is the only source of truth for the model.
The model can only answer using this text.
If the answer is not here, the model will guess with low confidence.

```python

context = """
This course introduces students to natural language processing techniques.
Topics include tokenization, sentiment analysis, information extraction,
machine translation, and question answering. The course uses Python
and popular NLP libraries such as NLTK, spaCy, and Hugging Face.
"""
```

```python

questions = [
    "What programming language is used in this course?",
    "Which libraries are used for NLP in this course?",
    "Who is the instructor of this course?",
    "What do you know about group number 3?",
    "Where is this course being taken?"
]
```

The model scans the context
It looks for the best text span that answers the question
It returns:

answer: extracted text

score: confidence level (0 to 1)

```python

# 1- answer: extracted text span
# 2- score: confidence (higher usually means more confident) Sometimes it also returns start/end positions in the context.

rows = []

for q in questions:
    result = qa(question=q, context=context)

    rows.append({
        "question": q,
        "answer": result["answer"],
        "score": result["score"]
    })
```

You store:

The question

The modelâ€™s answer

How confident the model is

```python
qa_df = pd.DataFrame(rows)
```

```python
# Add a new column based on the score
qa_df["score_status"] = qa_df["score"].apply(
    lambda x: "Not found in context" if x < 0.20 else "Answer found"
)
```

```python
from transformers import pipeline #importing the pipeline function from the transformers library
import pandas as pd #importing the pandas library

qa = pipeline("question-answering") #starting a question-answering pipeline

context = "here" #defining the context passage
questions = "here" #defining the list of questions

rows = [] #initializing an empty list to store the results

for q in questions: #iterating over each question in the list of questions
    result = qa(question=q, context=context) #getting the answer and score from the QA pipeline

    rows.append({ #appending the question, answer, and score to the rows list
        "question": q, #the question being asked
        "answer": result["answer"], #the answer extracted by the QA model
        "score": result["score"] #the confidence score of the answer
    }) #end of appending to rows

qa_df = pd.DataFrame(rows) #creating a DataFrame from the rows list
qa_df["score_status"] = qa_df["score"].apply( #adding a new column based on the score
    lambda x: "Not found in context" if x < 0.25 else "Answer found" #definig the threshold for score status 0,25
) #end of applying the function to create score_status column

qa_df #displaying the final DataFrame with questions, answers, scores, and score status
```
