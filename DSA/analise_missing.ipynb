```python
import pandas as pd 
import missingno as msno
import matplotlib.pyplot as plt

dados = pd.read_csv("new_data.csv", sep=",", encoding="utf-8")
```

```python
msno.matrix(dados, figsize = (10, 6), sparkline = False)
plt.show()
```

```python
# Calcula a quantidade total de valores por coluna
total_valores = len(dados)
faltando = dados.isnull().sum()
percentual = (faltando / total_valores) * 100

# Cria um DataFrame com as três colunas
resultado = pd.DataFrame({
    'Total': total_valores,
    'Valores Ausentes': faltando,
    'Percentual de Ausência': percentual
})

# Exibe o resultado
print(resultado)
```
