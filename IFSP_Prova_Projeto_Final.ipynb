```
#Desenvolver um projeto de LANÇAMENTO DE NOTAS:
#Crie uma estrutura que permita o lançamento de notas para varios alunos
#Lançamento Nota, Nome e Sexo

#Cada aluno terá 03 notas. Observando:
  #Nenhuma poderá receber notas inferiores a 0(zero) 
  #Nenhuma poderá receber notas superior a 10(dez)   

#Calcule a media do aluno.
  #Somatorio das 3 notas e divida por 3              

  #Media >= 7 APROVADO                               
  #Media entre >= 4 e > 7 EXAME                      
  #Media < 4 REPROVADO                               

#Ao final apresente:
  #Total de alunos cadastrados                       

  #PORCENTAGEM
    #Quantidade percentual de alunos APROVADOS
    #Quantidade percentual de alunos de EXAME
    #Quantidade percentual de alunos REPROVADOS

  #VALORES ABSOLUTOS
    #Quantidade de pessoas do sexo FEMININO APROVADAS
    #Quantidade de pessoas do sexo FEMININO EXAME
    #Quantidade de pessoas do sexo FEMININO REPROVADA

    #Quantidade de pessoas do sexo MASCULINO APROVADAS
    #Quantidade de pessoas do sexo MASCULINO EXAME
    #Quantidade de pessoas do sexo MASCULINO REPROVADA
```



````python
#TITULO DO PROJETO
print('INICIANDO PROJETO - LANÇAMENTO DE NOTAS')
print('\n')

#VARIAVEIS PARA CONTROLE
#QUANTIDADE TOTAL DE ALUNO
total_aluno = 0

#QUANTIDADE ALUNOS APROVADOS
qtde_aprovado_F = 0
qtde_aprovado_M = 0

#QUANTIDADE ALUNOS EXAME
qtde_exame_F = 0
qtde_exame_M = 0

#QUANTIDADE ALUNOS REPROVADOS
qtde_reprovado_F = 0
qtde_reprovado_M = 0


#VALIDAÇÃO PARA INICIAR O SISTEMA
inserir_aluno = input('Deseja Inserir Um Novo Aluno ? [S] ou [N] ')

print('\n')

#RESPOSTA SIM
while inserir_aluno.upper() == 'S':
  #NOME DO ALUNO
  nome_aluno = input('Digite nome do aluno (a):')
  sexo_aluno = input('Digite o sexo do aluno (a): F ou M ')
  while sexo_aluno.upper() != 'F' and sexo_aluno.upper() != 'M':
    print('opção invalida, digite sexo: F ou M')
    sexo_aluno = input('Digite o sexo do aluno (a): F ou M ')

  if sexo_aluno.upper() == 'F':
    total_aluno +=1
  elif sexo_aluno.upper() == 'M':
    total_aluno +=1
  
  #NOTA DO ALUNO
  n1 = int(input('Digite 1ª nota do aluno: '))
  while n1 <0 or n1>10:
    print('Nota invalida, deve ser entre 1 a 10')
    n1 = int(input('Digite 1ª nota do aluno: '))
  
  n2 = int(input('Digite 2ª nota do aluno: '))
  while n2 <0 or n2>10:
    print('Nota invalida, deve ser entre 1 a 10')
    n2 = int(input('Digite 2ª nota do aluno: '))
  
  n3 = int(input('Digite 3ª nota do aluno: '))
  while n3 <0 or n3>10:
    print('Nota invalida, deve ser entre 1 a 10')
    n3 = int(input('Digite 3ª nota do aluno: '))
 
  media = (n1+n2+n3)/3

#PARA ALUNOS APROVADOS
  if media >=7 and sexo_aluno.upper() == 'F':
   qtde_aprovado_F +=1

  elif media >=7 and sexo_aluno.upper() == 'M':
   qtde_aprovado_M += 1

#PARA ALUNOS DE EXAME
  elif media >=4 and sexo_aluno.upper() == 'F':
   qtde_exame_F +=1

  elif media >=4 and sexo_aluno.upper() == 'M':
   qtde_exame_M += 1
  
#PARA ALUNOS REPROVADOS
  elif media < 4 and sexo_aluno.upper() == 'F':
   qtde_reprovado_F +=1

  elif media < 4 and sexo_aluno.upper() == 'M':
   qtde_reprovado_M += 1

  
  inserir_aluno = input('Deseja Inserir Um Novo Aluno ? [S] ou [N] ')


#RESPOSTA DIFERENTE
while inserir_aluno.upper() != 'S' and inserir_aluno.upper() != 'N':
  print('Alternativa incorreta, preencha [S] ou [N]')
  inserir_aluno = input('Deseja Inserir Um Novo Aluno ? [S] ou [N] ')

  while inserir_aluno.upper() == 'S':
    #NOME DO ALUNO
    nome_aluno = input('Digite nome do aluno (a):')
    sexo_aluno = input('Digite o sexo do aluno (a): F ou M ')
    while sexo_aluno.upper() != 'F' and sexo_aluno.upper() != 'M':
      print('opção invalida, digite sexo: F ou M')
      sexo_aluno = input('Digite o sexo do aluno (a): F ou M ')

    if sexo_aluno.upper() == 'F':
      total_aluno +=1
    elif sexo_aluno.upper() == 'M':
      total_aluno +=1
    
    #NOTA DO ALUNO
    n1 = int(input('Digite 1ª nota do aluno: '))
    while n1 <0 or n1>10:
      print('Nota invalida, deve ser entre 1 a 10')
      n1 = int(input('Digite 1ª nota do aluno: '))
    
    n2 = int(input('Digite 2ª nota do aluno: '))
    while n2 <0 or n2>10:
      print('Nota invalida, deve ser entre 1 a 10')
      n2 = int(input('Digite 2ª nota do aluno: '))
    
    n3 = int(input('Digite 3ª nota do aluno: '))
    while n3 <0 or n3>10:
      print('Nota invalida, deve ser entre 1 a 10')
      n3 = int(input('Digite 3ª nota do aluno: '))
  
    media = (n1+n2+n3)/3

  #PARA ALUNOS APROVADOS
    if media >=7 and sexo_aluno.upper() == 'F':
      qtde_aprovado_F +=1

    elif media >=7 and sexo_aluno.upper() == 'M':
      qtde_aprovado_M += 1

  #PARA ALUNOS DE EXAME
    elif media >=4 and sexo_aluno.upper() == 'F':
      qtde_exame_F +=1

    elif media >=4 and sexo_aluno.upper() == 'M':
      qtde_exame_M += 1
    
  #PARA ALUNOS REPROVADOS
    elif media < 4 and sexo_aluno.upper() == 'F':
      qtde_reprovado_F +=1

    elif media < 4 and sexo_aluno.upper() == 'M':
      qtde_reprovado_M += 1

    
    inserir_aluno = input('Deseja Inserir Um Novo Aluno ? [S] ou [N] ')



print('PROGRAMA FINALIZADO')
print('\n') 
print('INFORMAÇÔES: \n')

if total_aluno >0:
  #TOTAL DE ALUNOS CADASTRADOS
  print('Quantidade total de alunos CADASTRADOS: {} alunos'.format(total_aluno))
  print('\n')

  #PERCENTUAIS
  print('Quantidade percentual de alunos APROVADOS: {:.1f}% alunos'.format(((qtde_aprovado_F + qtde_aprovado_M)/total_aluno)*100))
  print('Quantidade percentual de alunos EXAME: {:.1f}% alunos'.format(((qtde_exame_F + qtde_exame_M)/total_aluno)*100))
  print('Quantidade percentual de alunos REPROVADOS: {:.1f}% alunos'.format(((qtde_reprovado_F + qtde_reprovado_M)/total_aluno)*100))
  print('\n')

  #VALORES ABSOLUTOS
  #FEMININO
  print('Quantidade de pessoas do sexo FEMININO APROVADAS: {} pessoas'.format(qtde_aprovado_F))
  print('Quantidade de pessoas do sexo FEMININO EXAME: {} pessoas'.format(qtde_exame_F))
  print('Quantidade de pessoas do sexo FEMININO REPROVADA: {} pessoas'.format(qtde_reprovado_F))
  print('\n')

  #MASCULINO
  print('Quantidade de pessoas do sexo MASCULINO APROVADAS: {} pessoas'.format(qtde_aprovado_M))
  print('Quantidade de pessoas do sexo MASCULINO EXAME: {} pessoas'.format(qtde_exame_M))
  print('Quantidade de pessoas do sexo MASCULINO REPROVADA: {} pessoas'.format(qtde_reprovado_M)) 
  print('\n')

else:
  print('Você não inseriu nenhum aluno')
  print('Quantidade total de alunos CADASTRADOS: {} alunos'.format(total_aluno))
````
