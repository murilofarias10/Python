```python
#Projeto Final IFPS_SERTÃOZINHO

#Desenvolver um projeto de LANÇAMENTO DE NOTAS:

#Crie uma estrutura que permita o lançamento de notas para varios alunos

#Lançamento Nota, Nome e Sexo

#Cada aluno terá 03 notas. Observando:
  #Nenhuma poderá receber notas inferiores a 0(zero) ************
  #Nenhuma poderá receber notas superior a 10(dez)   ************

#Calcule a media do aluno.
  #Somatorio das 3 notas e divida por 3              ************

  #Media >= 7 APROVADO                               ************
  #Media entre >= 4 e > 7 EXAME                      ************
  #Media < 4 REPROVADO                               ************

#Ao final apresente:
  #Total de alunos cadastrados                       ************

  #PORCENTAGEM
    #Quantidade percentual de alunos APROVADOS
    #Quantidade percentual de alunos de EXAME
    #Quantidade percentual de alunos REPROVADOS

  #VALORES ABSOLUTOS
    #Quantidade de pessoas do sexo FEMININO APROVADAS
    #Quantidade de pessoas do sexo FEMININO EXAME
    #Quantidade de pessoas do sexo FEMININO REPROVADA

    #Quantidade de pessoas do sexo MASCULINO APROVADAS
    #Quantidade de pessoas do sexo MASCULINO EXAME
    #Quantidade de pessoas do sexo MASCULINO REPROVADA
```

```python
print('INICIANDO PROJETO - LANÇAMENTO DE NOTAS')
print('\n')


lista_nota = []
lista_nome = []
lista_media = []
inicio = 1
final = 4

masculino = 0
feminino = 0

qtde_aprovado = 0
qtde_reprovado = 0
qtde_exame = 0

inserir_aluno = input('Deseja Inserir Um Novo Aluno ? [S] ou [N] ')

print('\n')

#RESPOSTA SIM

while inserir_aluno.upper() == 'S':
  #NOME DO ALUNO
  nome_aluno = input('Digite nome do aluno (a):')
  sexo_aluno = input('Digite o sexo do aluno (a): F ou M ')
  while sexo_aluno.upper() != 'F' and sexo_aluno.upper() != 'M':
    print('opção invalida, digite sexo: F ou M')
    sexo_aluno = input('Digite o sexo do aluno (a): F ou M ')

  if sexo_aluno.upper() == 'F':
    feminino +=1
  elif sexo_aluno.upper() == 'M':
    masculino +=1

  lista_nome.append(nome_aluno)
  lista_nota = []
  #NOTA DO ALUNO
  inicio = 1
  final = 4
  while inicio < final:
    nota = int(input('Digite '+ str(inicio)+ 'ª nota do aluno: ' + nome_aluno + ' '))
    while nota < 0 or nota > 10:
      print('Valor da nota deve ser maior que 0 e menor que 10')
      nota = int(input('Digite '+ str(inicio)+ 'ª nota do aluno: ' + nome_aluno + ' '))
    
    lista_nota.append(nota)
    inicio +=1
  media = (lista_nota[0] + lista_nota[1] + lista_nota[2]) / 3
  lista_media.append(media)
  inserir_aluno = input('Deseja Inserir Um Novo Aluno ? [S] ou [N] ')
  
  
#RESPOSTA NAO
while inserir_aluno.upper() == 'N':
  print('fim do programa')
  break

#ALTERNATIVA INCORRETA
while inserir_aluno.upper() != 'S' and inserir_aluno.upper() != 'N':
  print('Alternativa incorreta, preencha [S] ou [N]')
  inserir_aluno = input('Deseja Inserir Um Novo Aluno ? [S] ou [N] ')
  
  if inserir_aluno.upper() == 'N':
   print('fim do programa')
   break

  if inserir_aluno.upper() == 'S':
  #NOME DO ALUNO
    nome_aluno = input('Digite nome do aluno (a):')
    sexo_aluno = input('Digite o sexo do aluno (a): F ou M ')
    while sexo_aluno.upper() != 'F' and sexo_aluno.upper() != 'M':
        print('opção invalida, digite sexo: F ou M')
        sexo_aluno = input('Digite o sexo do aluno (a): F ou M ')

    if sexo_aluno.upper() == 'F':
        feminino +=1

    elif sexo_aluno.upper() == 'M':
        masculino +=1

    lista_nome.append(nome_aluno)
    lista_nota = []
    #NOTA DO ALUNO
    inicio = 1
    final = 4
    while inicio < final:
      nota = int(input('Digite '+ str(inicio)+ 'ª nota do aluno: ' + nome_aluno + ' '))
      while nota < 0 or nota > 10:
        print('Valor da nota deve ser maior que 0 e menor que 10')
        nota = int(input('Digite '+ str(inicio)+ 'ª nota do aluno: ' + nome_aluno + ' '))
    
      lista_nota.append(nota)
      inicio +=1
    media = (lista_nota[0] + lista_nota[1] + lista_nota[2]) / 3
    lista_media.append(media)
    inserir_aluno = input('Deseja Inserir Um Novo Aluno ? [S] ou [N] ')
    if inserir_aluno.upper() == 'N':
      print('fim do programa')
      break
    
inicio = 1
total_nome = len(lista_nome) #quantidade de itens da lista nome
while inicio <= total_nome:
  valor = inicio -1
  resultado_1 = lista_nome[valor]
  resultado_2 = lista_media[valor]
  aprovacao = resultado_2
  if aprovacao >= 7:
      titulo = 'APROVADO'
      qtde_aprovado +=1
  elif aprovacao >=4 and aprovacao <7:
      titulo = 'EXAME'
      qtde_exame +=1
  else:
        titulo = 'REPROVADO'
        qtde_reprovado +=1

  print('\n')
  print('Nome: {} Média: {:.2f} Resultado: {}'.format(resultado_1,resultado_2,titulo))
            #print('Nome: ' + str(resultado_1) +  ' Media: ' + str(resultado_2)+ 'Resultado: ' + str(titulo))
  inicio +=1
print('\n')
print('Quantidade de pessoas do sexo masculino cadastrado é: {}'.format(masculino))
print('Quantidade de pessoas do sexo feminino cadastrado é: {}'.format(feminino))
print('Quantidade total de pessoas cadastradas é: {}'.format((len(lista_nome))))
print('\n')
print('Quantidade de pessoas aprovadas é: {} '.format(qtde_aprovado))
print('Quantidade de pessoas reprovadas é: {}'.format(qtde_reprovado))
print('Quantidade de pessoas exame é: {}'.format(qtde_exame))
  
```
