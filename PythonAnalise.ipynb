```python
#Importando Bibliotecas
import pandas as pd
import urllib
import matplotlib.pyplot as plt
import seaborn as sns
#Importando arquivos
arquivo = pd.read_excel(r'/content/drive/MyDrive/Colab Notebooks/BASEDADOSMRP.xlsx')
#mostrando informações do arquivo
display(arquivo.info())
#gerando grafico de pizza
import matplotlib.pyplot as plt
fig1, ax1 = plt.subplots()
area = arquivo['AREA'].value_counts()
legenda = arquivo['AREA'].value_counts()
explode = (0, 0.1, 0, 0)
ax1.pie(area, shadow=True, explode=explode, startangle=180, autopct='%1.1f%%',labels=legenda)
plt.show()
#geradno grafico wistia
sns.heatmap(arquivo.corr(), cmap="Wistia", annot=True)
#gerando graficos de colunas
import plotly.express as px
grafico = px.histogram(arquivo, x ="ID CWP", color="AREA")
grafico.show()
import plotly.express as px
grafico = px.histogram(arquivo, x ="ID PRIMAVERA", color="AREA")
grafico.show()
```
